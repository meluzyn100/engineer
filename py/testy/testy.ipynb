```python
Xt = [np.random.choice([0, 1], p=[1-p, p])]
t = 1 
p = 0.3
delta = 0.1
e = 0.01
LB = 0
UB = 2
while (1+e)*LB < (1-e)*UB:
    t += 1
    Xt.append(np.random.choice([0, 1], p=[1-p, p]))
    mu = np.mean(Xt)
    sig = np.sqrt(np.sum((Xt-mu)**2)/t)
    ct = sig*np.sqrt(2*np.log(3/delta)/t) + 3*np.log(3/delta)/t
    LB = max(LB,mu - ct)
    UB = min(UB, mu + ct)
    if t%1000 == 0:
        print(LB, UB, t)
```

```python
t = 1 
p = 0.5
delta = 0.1
e = 0.01
LB = 0
UB = 2
pol = 2
Xt = [np.random.choice([0, 1], p=[1-p, p])]
while (1+e)*LB < (1-e)*UB:
    t += 1
    Xt.append(np.random.choice([0, 1], p=[1-p, p]))
    mu = np.mean(Xt)
    sig = np.sqrt(np.sum((Xt-mu)**2)/t)
    c = delta*(pol -1)/pol
    dt = c/t**pol
    # df = delta
    ct = sig*np.sqrt(2*np.log(3/dt)/t) + 3*np.log(3/dt)/t
    LB = max(LB,mu - ct)
    UB = min(UB, mu + ct)
    if t%1000 == 0:
        print(LB, UB, dt, ct, t)
```
